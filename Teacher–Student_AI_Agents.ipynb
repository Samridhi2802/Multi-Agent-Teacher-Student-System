# **Project Overview:**

Provide a high-level overview of the Multi-Agent Teacher-Student System.

The **Multi-Agent Teacher-Student System** is a project built using Google's Agent Development Kit (ADK) that simulates a dynamic learning environment. Its primary purpose is to demonstrate a collaborative and structured conversational flow between AI agents embodying distinct roles.

**Purpose and Functionality:**
This system aims to facilitate a multi-step reasoning process where knowledge is explained, summarized, and questioned. It features three main agents:

*   **Teacher Agent:** Acts as a senior professor, designed to provide clear and comprehensive explanations of a given topic.
*   **Student Agent:** Summarizes the teacher's explanation, identifies key points, and formulates follow-up questions to deepen understanding or clarify concepts.
*   **Root Agent:** Serves as the orchestrator, managing the interaction between the Teacher and Student agents. It takes the initial user query, passes it to the Teacher, then directs the Teacher's response to the Student, and finally returns the Student's processed output to the user.

**Problem Solved:**
The system addresses the challenge of creating interactive and intelligent educational or explanatory tools. Instead of a single AI providing a static response, it simulates a more natural, multi-faceted learning interaction. This approach allows for:

*   **Structured Information Delivery:** Ensuring concepts are broken down and presented systematically.
*   **Active Learning Simulation:** Encouraging critical thinking through summarization and questioning.
*   **Adaptive Engagement:** The chained interaction means the system dynamically responds to the previous output, mimicking a more engaging and responsive tutor.

**General Approach:**
The core of the system relies on the Google ADK to define and connect Large Language Model (LLM) based agents. Each agent is configured with a specific role, name, description, and set of instructions that guide its behavior. The `root_agent` leverages a `teacher_student_chain` function as a 'tool' to sequentially invoke the `teacher` and `student` agents, processing information in a pipeline. This setup effectively creates a sophisticated AI that can provide explanations and engage in follow-up dialogue, offering a more nuanced interactive experience.

### 2. `get_agent_response_text` Helper Function

```python
async def get_agent_response_text(agent_instance: Agent, message: str) -> str:
    full_response_text = ""
    async for event in agent_instance.run_async(None, message):
        if hasattr(event, 'text'):
            full_response_text += event.text
    return full_response_text
```

-   This asynchronous function is crucial for extracting text responses from ADK agents.
-   ADK agents' `run_async` method yields an asynchronous generator of event objects, not direct strings.
-   This helper iterates through these events, checks if an event has a `text` attribute, and concatenates all text fragments into a single string.
-   `None` is passed as the first argument to `run_async` to explicitly indicate no `parent_context` is being used, addressing a specific API requirement.

### 3. `teacher_student_chain` Function

```python
async def teacher_student_chain(message: str):
    t = await get_agent_response_text(teacher, message)
    s = await get_agent_response_text(student, t)
    return s
```

-   This asynchronous function defines the sequential interaction: The `teacher` agent first responds to the initial `message`.
-   The teacher's response (`t`) is then passed to the `student` agent, which processes it and generates its own response (`s`).
-   This effectively creates a chained conversation flow.

**In Short:** This helper iterates through these events, checks if an event has a text attribute, and concatenates all text fragments to form a complete response string as ADK agents don't return simple strings directly when run; instead, agent_instance.run_async(None, message) returns an asynchronous generator of event objects.

### 4. Connecting the Chain to the `root_agent` :

- The root_agent is given the teacher_student_chain function as a tool.
- In ADK, tools are functions that an agent can call to perform specific tasks.

When the root_agent is prompted, it will invoke this teacher_student_chain to produce a combined response from the teacher  and student agents.

````python
%%writefile teacher_student_agent/agent.py
from google.adk.agents.llm_agent import Agent
# Removed: from google.adk.events import TextEvent

teacher = Agent(
    model="gemini-2.0-flash-lite",
    name="teacher",
    description="Senior professor explaining AI.",
    instruction="Explain clearly like a professor."
)

student = Agent(
    model="gemini-2.0-flash-lite",
    name="student",
    description="Student summarizing and asking questions.",
    instruction="Summarize and ask a follow-up question."
)

# Helper function to get text response from an agent's run_async
async def get_agent_response_text(agent_instance: Agent, message: str) -> str:
    full_response_text = ""
    async for event in agent_instance.run_async(None, message): # Pass None for parent_context
        # Check if the event object has a 'text' attribute
        if hasattr(event, 'text'):
            full_response_text += event.text
    return full_response_text

# ADK requires a SINGLE root agent
# So we chain teacher â†’ student inside the root agent

async def teacher_student_chain(message: str):
    t = await get_agent_response_text(teacher, message)
    s = await get_agent_response_text(student, t)
    return s

root_agent = Agent(
    model="gemini-2.0-flash-lite",
    name="root_agent",
    description="Teacher & Student multi-agent conversation.",
    instruction="Act as a multi-step teacher-student reasoning engine.",
    tools=[teacher_student_chain]
)
````

## Encountered Errors and Their Solutions

During the development and testing of this multi-agent system, several common issues related to the `google-adk` library and `LlmAgent` interaction were encountered and resolved:

### 1. `AttributeError: 'LlmAgent' object has no attribute 'predict'`

-   **Problem:** An initial attempt to interact with the `teacher` and `student` agents used `await teacher.predict(message)`. However, `LlmAgent` instances in `google-adk` do not expose a direct `predict` method for simple text output in this manner.
-   **Solution:** The correct way to get a response from an `LlmAgent` is to call its `run_async` method, which returns an asynchronous generator of events. A helper function, `get_agent_response_text`, was introduced to iterate through these events and extract the text content.

### 2. `TypeError: 'LlmAgent' object is not callable`

-   **Problem:** After removing `predict`, a subsequent attempt was to call the `LlmAgent` instance directly like a function (e.g., `await teacher(message)`).
-   **Solution:** This error confirmed that `LlmAgent` objects are not directly callable. The `get_agent_response_text` helper function correctly uses `agent_instance.run_async()` to interact with the agent.

### 3. `ImportError: cannot import name 'TextEvent' from 'google.adk.events'`

-   **Problem:** The `google-adk` library's internal structure changed, leading to `TextEvent` no longer being directly importable from `google.adk.events`.
-   **Solution:** The `get_agent_response_text` helper function was updated to be more robust. Instead of relying on a specific `TextEvent` class, it now checks if any `event` object yielded by `run_async` has a `text` attribute (`if hasattr(event, 'text')`), allowing it to extract text regardless of the specific event type.

### 4. `AttributeError: 'str' object has no attribute 'model_copy'` (related to `run_async` argument order)

-   **Problem:** When calling `run_async`, the `message` string was sometimes incorrectly interpreted as a `parent_context` object, leading to an `AttributeError` when `google-adk` tried to call `model_copy()` on it. This indicated a change in the expected argument order for `run_async`.
-   **Solution:** The `get_agent_response_text` function was modified to explicitly pass `None` as the first argument for `parent_context` (i.e., `agent_instance.run_async(None, message)`). This ensures the `message` is correctly passed as the second argument, as intended.
